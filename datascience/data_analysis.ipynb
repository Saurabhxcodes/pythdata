# selecting rows and cols
rows (slicing)
loc - gets us the data using named indexes
iloc - gets us the data using numeric indexes
cols (dictionary style)
df[single] - requires name of column
df[multiple] - requires a list of columns (multiple is list of named cols)

```python
# single row (record)
weather.iloc[0]
```

```python
# records 25 - 50
weather.iloc[25: 50]
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, :5] 
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, [0,1,2,3,4]] 
```

```python
# 15 records with last 3 cols
weather.iloc[:15, -3:]
```

```python
# 15 records with last 3 cols
weather.iloc[:15, [-3,-2,-1]] # or weather.iloc[:15, [11,12,13]] 
```

```python
weather['Date'] # selecting a single column (dictionary style) -> Series
```

```python
cols = ['Date','MaxTemp','MinTemp']
weather[cols] # multiple cols (dictionary style) -> DataFrame
```
